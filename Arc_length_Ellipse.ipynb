```python
import math
import matplotlib.pyplot as plt
import plotly.express as px
from scipy.integrate import quad
```

```python
a_lst = [i/10 for i in range(20, 40) for j in range(10)]
b_lst = [i/10 for i in range(20, 40) for j in range(10)]
print(a_lst,b_lst)
print(len(a_lst), len(b_lst))
```

**ARC LENGTH OF ELLIPSE**

```python
f_lst = [] # taking accumulator to store value of integration in list
def func(x, index):
  """
  This function integrate by using scipi
  """
  return ((a_lst[index]*math.sin(x))**2 + (b_lst[index]*math.cos(x))**2)**(0.5)
for i in range(200):
  result,error = quad(func,0,2*math.pi,args=(i))
  f_lst.append(result)
  # print(f_lst[i])
  # print(2*math.pi*a_lst[i])

```

**DRAWING PECENTAGE ERROR TO COMPARING GRAPH OF ARITHMATIC, GEOMATRIC AND RAMANUJAN APPROXIMATION**

```python
# Perimeter of Ellipse
a_lst = [i/10 for i in range(20, 40) ]
# b_lst = [i/10 for i in range(20, 40) ]
f_lst = []
def func(x, index):
  return ((a_lst[index]*math.sin(x))**2 + (1*math.cos(x))**2)**(0.5)
for i in range(20):
  result,error = quad(func,0,2*math.pi,args=(i))
  f_lst.append(result)
  
percentage_error_arithmatic_lst = []
percentage_error_geometric_lst = []
percentage_error_raman_lst = []
for i in range(20):
  h = ((a_lst[i] - 1)/(a_lst[i] + 1))**2
  perimeter_arithmatic = 2*math.pi*(a_lst[i] + 1)/2
  perimeter_raman = abs(math.pi*(a_lst[i]+1)*(1 + 3*h/(10*(4 - 3*h)**(0.5))))
  perimeter_geometric = 2*math.pi*((a_lst[i]*1)**(0.5))
  percentage_error_arithmatic = abs((f_lst[i] - perimeter_arithmatic)*100/f_lst[i])
  percentage_error_geometric = abs((f_lst[i] - perimeter_geometric)*100/f_lst[i])
  percentage_error_raman = ((f_lst[i] - perimeter_raman)*100/f_lst[i])
  percentage_error_arithmatic_lst.append(percentage_error_arithmatic)
  percentage_error_geometric_lst.append(percentage_error_geometric)
  percentage_error_raman_lst.append(percentage_error_raman)
```

```python
plt.plot(a_lst, percentage_error_arithmatic_lst , label = 'Arithmatic')
plt.legend()
plt.plot(a_lst, percentage_error_geometric_lst , label = 'Geometric')
plt.legend()
plt.plot(a_lst, percentage_error_raman_lst, label= "Ramanujan's approximation")
plt.legend()
```
