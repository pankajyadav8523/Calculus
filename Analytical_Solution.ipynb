```python
import math
import matplotlib.pyplot as plt
```

```python
# t is a variable which vary through x axis in interval [0,5]
t= []
dt = 0.1
for i in range(0,51):
  t.append(i/10)
```

```python
# analitical solution for x(t) = e**t

anl_sol = []
for i in t:
  x = math.exp(i)
  anl_sol.append(x)

plt.plot(t, anl_sol)
plt.show()
```

```python
# Numerical solution (Euler method) x1(t + dt) = dt*x1(t) + x1(t)
num_sol = [1]
dt = 0.1
for j in range(50):
    
  x1 = num_sol[j]*dt + num_sol[j]
    
  num_sol.append(x1)
plt.plot(t, anl_sol, label= 'Analytical solution')
plt.legend()
plt.plot(t, num_sol, label = 'Solution By Euler')
plt.legend()


```

```python
# error measurment between analytical and numerical method
#error = mod of (analitical solution - numerical solytion)/analitical solution
dt = []
for i in range(51):
  dt.append(i/1000)

# to find error
error = []
for j in range(51):
  res = (anl_sol[j] - num_sol[j])/anl_sol[j]
  error.append(res)

plt.plot(dt, error)
plt.show()
```

```python
# Modified Euler method (midpoint method) x1 = x0 + dt(x0 + dt*x0/2)
num_mid = [1]
dt = 0.1
for j in range(50):
  x1 = num_mid[j] + dt*(num_mid[j] + dt*num_mid[j]/2)
  num_mid.append(x1)
plt.plot(t, num_mid)
plt.show
```

```python
#measure error between Euler(midpoint method) and analytical method
dt = []
for i in range(51):
  dt.append(i/1000)

# to find error
error = []
for j in range(51):
  res = abs(anl_sol[j] - num_mid[j])/num_mid[j]
  error.append(res)

plt.plot(dt, error)
plt.show()
```
