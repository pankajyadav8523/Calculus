$$ GRADIENT \quad DESCENT $$


$$θ = θ - α\cdot ∇f(θ) $$

where:
>$* \quad θ$ represents the vector of parameters of model.

> $* \quad α$ is learning rate which controls the size of the step taken in the direction of the negative gradient.

> $* \quad f(θ)$ represents the gradient of the cost function $f$ with respect to $θ$.
Calculating minima for the function given below by using gradient descent:
$$f(x) = sin(x)$$
For this function in one variable gradient descent will updated as follows:
$$x_{n+1} = x_{n} - \alpha \frac{\partial f}{\partial x}$$
where:


> $* \quad x_{n}$ is starting guessing point on function.

Others refer to same meaning as above mentioned.

```python
# Symbolic method to find derivative of sin(x)
# for example - d(sin(x))/dx = cos(x)

x = [(i*math.pi/10 + 0.1*i) for i in range(90)]
y1 = [math.cos(x[i]) for i in range(90)]
```

```python
# Numerical method to find derivative of sin(x)
# dy/dx = sin(x + dx) - sin(x)/dx

dx = 0.9
y2 = [(math.sin(x[i] + dx) - math.sin(x[i]))/dx for i in range(90)]
```

```python
# Autograde method to find derivative of sin(x)

import autograd.numpy as np
from autograd import grad
def my_function(x):
  return np.sin(x)
# Find the gradient of my_function
gradient_func= grad(my_function)
y3 = []
for i in range(90):
  gradient = gradient_func(x[i])
  y3.append(gradient)
```

```python
plt.plot(x, y1, label= 'Symbolic')
plt.legend()
plt.plot(x, y2, label = 'Numrical')
plt.legend()
plt.plot(x, y3, label = 'Autograde')
plt.legend()
```

```python
# Pseaudo code for using gradient descent to find minima 

# first find out the derivative of loss or cost function by using Autograd method
# use formula of gradient descent C(new) = C(old)  - (LR*slope) 
# Iterate on the gredient descent
# set stopping threshold where  we want to stop iteraton

# Gradient descent for one variable

import math
import autograd.numpy as np
from autograd import grad

eta = 0.01 # This is a Learning rate
x_new = [25.0] # it is the old value or initialiy randomly chosen point
def cost_derivative(x):
  """
  Returns: it return the slope of function
  Parameter x: it is variable where we want to w.r.t differentiate
  """
  return (x-2)**2
slope = grad(cost_derivative) # To find slope by using autograd method

def display_minima(x):
  """
  It display the value of minima and slope at minima
  Parameter x: list of x-coordinate
  """

  for i in x_new: # Iterate over the x- axis 
    x1 = i - eta*slope(i)
    x_new.append(x1)
    if abs(slope(i)) < 1e-3: # it Iterates until the slope tends to Zero(because at the point of minima slope is equal to zero)
      print(i, slope(i))
      break

display_minima(x_new)
```

GRADIENT DESCENT FOR TWO VARIABLE

```python
# Pseaudo code for using gradient descent to find minima 

# first find out the derivative of loss or cost function by using Autograd method
# use formula of gradient descent C(new) = C(old)  - (LR*slope) 
# Iterate on the gredient descent
# set stopping threshold where  we want to stop iteraton

# Gradient descent for two variable

import autograd.numpy as np
from autograd import grad

def my_function(X):
  """
  Return: x ** 2 + y ** 2 function.

  Parameter X: It is a argument given to function.
  Precondition: It is any argument.
  """
  x,y = X
  return x**2 + y**2
gradient = grad(my_function)
```

```python
# calculating minima.
x_n = [float(input(f"enter value {i}:")) for i in range(1, 3)]             # initially randomly choosen point.
eta = 0.001                # learning rate.
previous = [0.0, 0.0]   # accumulator to keep previous values track
val = True
while val:
    x_n[0] = x_n[0] - eta*(gradient(x_n)[0]) # gradient decent for first variable
    x_n[1] = x_n[1] - eta*(gradient(x_n)[1]) # gradient decent for second variable
    if abs(gradient(x_n)[0] - gradient(previous)[0])<1e-3 and abs(gradient(x_n)[1] - gradient(previous)[1])<1e-3:
        val = False
        print(x_n)
    previous = x_n.copy()    # avoiding loss of values
     
```
